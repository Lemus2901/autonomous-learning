# 🧩 BLOQUE 1 — Manejo de valores nulos

## 🎯 Problema

Los datos faltantes rompen análisis y modelos.

---

## 🏋️ Ejercicios

1️⃣ Mostrar el monto y si es NULL poner 0
2️⃣ Mostrar la ciudad y si es NULL poner "DESCONOCIDA"
3️⃣ Mostrar fecha y si es NULL poner "1900-01-01"
4️⃣ Calcular gasto total por cliente tratando NULL como 0

---

## ✅ Ejemplos resueltos

### Ejemplo 1

```sql
SELECT 
    transaction_id,
    COALESCE(amount, 0) AS amount_limpio
FROM transactions;
```

---

### Ejemplo 2

```sql
SELECT 
    customer_id,
    COALESCE(city, 'DESCONOCIDA') AS city_limpia
FROM customers;
```

---

### Ejemplo 3

```sql
SELECT 
    customer_id,
    SUM(COALESCE(amount,0)) AS gasto_total
FROM transactions
GROUP BY customer_id;
```

# 🧩 BLOQUE 2 — Limpieza de texto

## 🎯 Problema

Errores de escritura dañan reportes y agrupaciones.

Ejemplo real:

```
"medellin"
"Medellin "
" MEDELLIN"
```

Son la misma ciudad pero SQL las ve diferentes.

---
## 🏋️ Ejercicios

1️⃣ Convertir nombres a **MAYÚSCULAS**  
2️⃣ Convertir nombres a **minúsculas**  
3️⃣ Quitar espacios al inicio y al final  
4️⃣ Reemplazar `"medellin"` por `"MEDELLÍN"`  
5️⃣ Reemplazar `"bogota"` por `"BOGOTÁ"`  
6️⃣ Quitar espacios internos dobles

---

## ✅ Ejemplos resueltos

### Ejemplo 1 — Mayúsculas

```sql
SELECT 
    UPPER(name) AS name_clean
FROM customers;
```

---

### Ejemplo 2 — Quitar espacios

```sql
SELECT 
    TRIM(name) AS name_clean
FROM customers;
```

---

### Ejemplo 3 — Reemplazar texto

```sql
SELECT 
    REPLACE(city, 'medellin', 'MEDELLÍN') AS city_clean
FROM customers;
```

---

### Ejemplo 4 — Todo junto (industria real)

```sql
SELECT
    customer_id,
    UPPER(TRIM(name)) AS name_clean,
    UPPER(TRIM(city)) AS city_clean
FROM customers;
```

# 🧩 BLOQUE 3 — Limpieza y formato de fechas

## 🎯 Problema

Las fechas mal formateadas impiden análisis por mes, año, etc.

---

## 🏋️ Ejercicios

1️⃣ Extraer solo el año de una fecha
2️⃣ Extraer mes
3️⃣ Agrupar ventas por año
4️⃣ Agrupar ventas por mes
5️⃣ Mostrar fecha como "YYYY-MM"
6️⃣ Calcular antigüedad del cliente

---

## ✅ Ejemplos resueltos

### Ejemplo 1 — Extraer año

```sql
SELECT 
    EXTRACT(YEAR FROM date) AS año
FROM transactions;
```

---

### Ejemplo 2 — Agrupar por mes

```sql
SELECT 
    EXTRACT(MONTH FROM date) AS mes,
    SUM(amount) AS ventas
FROM transactions
GROUP BY mes;
```

---

### Ejemplo 3 — Formato año-mes

```sql
SELECT 
    DATE_TRUNC('month', date) AS año_mes,
    SUM(amount)
FROM transactions
GROUP BY año_mes;
```

---

### Ejemplo 4 — Antigüedad cliente

```sql
SELECT 
    customer_id,
    CURRENT_DATE - MIN(date) AS dias_como_cliente
FROM transactions
GROUP BY customer_id;
```
